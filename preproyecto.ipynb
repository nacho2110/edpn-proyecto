```python

def amp_polym(x: float, rc: float) -> float:
    """Función de amplificación cúbica."""
    return x*(1-x)*(x-rc)
x = np.linspace(0, 1, 1000)
plt.figure(figsize=(5, 5))
classifier = BinaryClassifier(x0, 1, 1000, 1)
plt.plot(x, [classifier._f(i) for i in x], label='Quantization')
```

```python
plt.plot(x, x + amp_polym(x, 0.5), label='Amplification')
```
