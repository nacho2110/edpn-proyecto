```python

def amp_polym(x: float, rc: float) -> float:
    """Funcin de amplificacin cbica."""
    return x*(1-x)*(x-rc)
x = np.linspace(0, 1, 1000)
plt.figure(figsize=(15, 15))
classifier = BinaryClassifier(x0, 1, 10, 1)
plt.scatter(x, [classifier._f(i) for i in x], label='Amplification (f)')
plt.scatter(x, [classifier.q(i) for i in x], label='Quantization (q)')
plt.legend()
```

```python
plt.plot(x, x + amp_polym(x, 0.5), label=' f(x) = x(1-x)(x-0.5)')
plt.plot(x, x, label='identidad')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0.5, color='red', linestyle='--', linewidth=0.5)
plt.axvline(0.5, color='red', linestyle='--', linewidth=0.5)
plt.title('Funcin de amplificacin cbica con rho = 0.5')
plt.legend()
plt.grid()
plt.savefig("amplification_plot.pdf", format='pdf')
```
